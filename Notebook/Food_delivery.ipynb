```python
#Calculate distances between the longitudes and latitudes using haversine formula

radius = 6371

def to_rad(degrees):
  return degrees * (np.pi/ 180)

def calculate_dist(lat1, lon1, lat2, lon2):
  lat_dif = to_rad(lat2 - lat1)
  lon_dif = to_rad(lon2 - lon1)
  a = np.sin(lat_dif/2)**2 + np.cos(to_rad(lat1)) * np.cos(to_rad(lat2)) * np.sin(lon_dif/2)**2
  c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
  return radius * c

df['distance'] = np.nan

for i in range(len(df)):
  df.loc[i, 'distance'] = calculate_dist(df.loc[i, 'Restaurant_latitude'],
                                         df.loc[i,'Restaurant_longitude'],
                                         df.loc[i, 'Delivery_location_latitude'],
                                         df.loc[i, 'Delivery_location_longitude'])
```

```python
x = np.array(df[["Delivery_person_Age",
                   "Delivery_person_Ratings",
                   "distance"]])

y = np.array(df[["Time_taken(min)"]])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.1, random_state=42)
```

```python
x.shape  , y.shape
```

```python
#Model

def create_lstm_model(input_shape):
  input_layer = Input(shape = input_shape)
  lstm_1 = LSTM(128, return_sequences= True)(input_layer)
  lstm_2 = LSTM(64, return_sequences= False)(lstm_1)
  dense_1 = Dense(25)(lstm_2)
  output_layer = Dense(1)(dense_1)
  model = Model(inputs = input_layer, outputs = output_layer)
  return model

input_shape = (x_train.shape[1], 1)

model = create_lstm_model(input_shape)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(x_train, y_train, batch_size=1, epochs = 10)
```

```python
model.save('Model_train.h5')
```

```python
print("Food Delivery Time Prediction")
a = int(input("Age of Delivery Partner: "))
b = float(input("Ratings of Previous Deliveries: "))
c = int(input("Total Distance: "))

features = np.array([[a, b, c]])
print("Predicted Delivery Time in Minutes = ", model.predict(features))
```
